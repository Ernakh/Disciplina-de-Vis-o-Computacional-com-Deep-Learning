# Representando conceitos de Visão Computacional com CIFAR-10

Conceito 1 - dataset de classificação de imagem

```python
import torch
import torchvision
import matplotlib.pyplot as plt

from torchvision import datasets, transforms

cifar10 = datasets.CIFAR10(
    root="./data",
    train=True,
    download=True
)

classes = cifar10.classes

print("Quantidade de imagens:", len(cifar10))
print("Classes do CIFAR-10:", classes)

indice_img1 = 0
indice_img2 = 1

img1, label1 = cifar10[indice_img1]
img2, label2 = cifar10[indice_img2]

plt.figure(figsize=(6, 3))

plt.subplot(1, 2, 1)
plt.imshow(img1)
plt.title(f"Imagem 1: {classes[label1]}")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(img2)
plt.title(f"Imagem 2: {classes[label2]}")
plt.axis("off")

plt.show()
```

Conceito 2 — Imagem como tensor

```python
to_tensor = transforms.ToTensor()

tensor_img1 = to_tensor(img1)
tensor_img2 = to_tensor(img2)

print("Formato da imagem 1 como tensor:", tensor_img1.shape)
print("Formato da imagem 2 como tensor:", tensor_img2.shape)
```

Conceito 3 — Rótulo ou classe da imagem

```python
print("Rótulo numérico da imagem 1:", label1)
print("Classe da imagem 1:", classes[label1])

print("Rótulo numérico da imagem 2:", label2)
print("Classe da imagem 2:", classes[label2])
```

Conceito 4 — Aumento por espelhamento horizontal

```python
flip_horizontal = transforms.RandomHorizontalFlip(p=1.0)

img1_flip = flip_horizontal(img1)
img2_flip = flip_horizontal(img2)

plt.figure(figsize=(8, 4))

plt.subplot(2, 2, 1)
plt.imshow(img1)
plt.title("Imagem 1 original")
plt.axis("off")

plt.subplot(2, 2, 2)
plt.imshow(img1_flip)
plt.title("Imagem 1 espelhada")
plt.axis("off")

plt.subplot(2, 2, 3)
plt.imshow(img2)
plt.title("Imagem 2 original")
plt.axis("off")

plt.subplot(2, 2, 4)
plt.imshow(img2_flip)
plt.title("Imagem 2 espelhada")
plt.axis("off")

plt.show()
```

Conceito 5 — Aumento por recorte aleatório

```python
recorte_aleatorio = transforms.RandomResizedCrop(
    size=(32, 32),
    scale=(0.5, 1.0),
    ratio=(0.75, 1.33)
)

img1_crop = recorte_aleatorio(img1)
img2_crop = recorte_aleatorio(img2)

plt.figure(figsize=(8, 4))

plt.subplot(2, 2, 1)
plt.imshow(img1)
plt.title("Imagem 1 original")
plt.axis("off")

plt.subplot(2, 2, 2)
plt.imshow(img1_crop)
plt.title("Imagem 1 com recorte")
plt.axis("off")

plt.subplot(2, 2, 3)
plt.imshow(img2)
plt.title("Imagem 2 original")
plt.axis("off")

plt.subplot(2, 2, 4)
plt.imshow(img2_crop)
plt.title("Imagem 2 com recorte")
plt.axis("off")

plt.show()
```

Conceito 6 — Aumento por alteração de cor

```python
alteracao_cor = transforms.ColorJitter(
    brightness=0.5,
    contrast=0.5,
    saturation=0.5,
    hue=0.1
)

img1_cor = alteracao_cor(img1)
img2_cor = alteracao_cor(img2)

plt.figure(figsize=(8, 4))

plt.subplot(2, 2, 1)
plt.imshow(img1)
plt.title("Imagem 1 original")
plt.axis("off")

plt.subplot(2, 2, 2)
plt.imshow(img1_cor)
plt.title("Imagem 1 com alteração de cor")
plt.axis("off")

plt.subplot(2, 2, 3)
plt.imshow(img2)
plt.title("Imagem 2 original")
plt.axis("off")

plt.subplot(2, 2, 4)
plt.imshow(img2_cor)
plt.title("Imagem 2 com alteração de cor")
plt.axis("off")

plt.show()
```

## Conclusão

Nesta atividade, foi utilizado o dataset CIFAR-10 para representar seis conceitos presentes no capítulo de Visão Computacional do livro Mergulhe no aprendizado profundo.

Foram escolhidas duas imagens do dataset e, a partir delas, foram demonstrados os conceitos de dataset de classificação de imagem, imagem como tensor, rótulo/classe, aumento por espelhamento horizontal, aumento por recorte aleatório e aumento por alteração de cor.

A atividade mostra que, em visão computacional, as imagens podem ser manipuladas computacionalmente para gerar novas variações. Essas transformações são importantes porque aumentam a diversidade dos dados de treinamento e podem contribuir para que modelos de aprendizado profundo generalizem melhor.
